# ENSP Prerequisites

This is a proof-of-concept for using the networkx library to create a chart of prerequisites for the ENSP EMD program.

```python
%matplotlib inline
import networkx as nx
```

```python
# create directed graph
DG = nx.DiGraph()

# add nodes and edges for prerequisites
DG.add_edge('MATH 160', 'PHYS 210A')
DG.add_edge('MATH 160', 'MATH 161')
DG.add_edge('MATH 161', 'MATH 211S')
DG.add_edge('PHYS 114', 'ENSP 338')
DG.add_edge('PHYS 210A', 'ENSP 338')
DG.add_edge('PHYS 210A', 'ENSP 337')
DG.add_edge('PHYS 210A', 'PHYS 210B')
DG.add_edge('ENSP 337', 'ENSP 437')
DG.add_edge('ENSP 337', 'ENSP 439L')
DG.add_edge('ENSP 338','ENSP 438')
 
# draw graph (arrowheads are the thick lines)
nx.draw_circular(DG)

# todo: figure out how to plot in a more directed order
```
