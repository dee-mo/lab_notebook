# Price Duration Curve

This notebook generates graphs of a price duration curve corresponding to demand data from microgrids.  From the generator specification of fuel use rate in liters per hour and percent of rated load output power, we create a linear model of fuel use.  Since much of the observed demand is below the lowest point on the specification, we use a linear extrapolation.  From this we calculate the liters per kWh and calculate a price duration curve which shows the per kWh price for power against the duration. 

```python
%matplotlib inline
import numpy as np

import sys
sys.path.append('/Users/dsoto/Dropbox/Haiti-Soto/load-duration-curve/')
from ldc import Generator, plots, power_to_ldc
```

# Coteaux Kohler Generator

- from data sheet in haiti folder
- coteaux kohler
- percent load, standby lph, prime lph
- 100%, 44.3, 40.6
- 75%, 35.1, 32.3
- 50%, 26.3, 24.0
- 25%, 16.2, 14.4
- no load consumption not given
- from report: rated active power 125kW
- Manufactured 1994. 


```python
# create a generator object using the data from the Coteaux spec sheet
power_kW = 125.
efficiency_curve = {'output_power': np.array([0.25, 0.5, 0.75, 1.0]) * power_kW,
                    'fuel_use' : np.array([14.4, 24.0, 32.3, 40.6])}
coteaux = Generator(efficiency_curve)


power = np.loadtxt('/Users/dsoto/Dropbox/Haiti Micro-Grid Research/Generator output data/Coteaux/Full year time series/coteaux_power_tot.csv')
ldc = power_to_ldc(power)
plots(coteaux, ldc, 'coteaux')
```

# Port-A-Piment FGWilson Generator

- portapiment FGWilson
- @50Hz percent load, standby lph, prime lph
- 110%, -, 54.3
- 100%, 45.3, 49.4
- 75%, 41.0, 37.4
- 50%, 28.0, 25.6
- from report: rated active power 199 kW
- Manufactured 2009.


```python
# produce plots for port-a-piment data and generator specifications
power_kW = 199.
efficiency_curve = {'output_power': np.array([0.5, 0.75, 1.0, 1.1]) * power_kW,
                    'fuel_use' : np.array([25.6, 37.4, 49.4, 54.3])}

# create generator with characteristics from data sheet
pap = Generator(efficiency_curve)
power = np.loadtxt('/Users/dsoto/Dropbox/Haiti Micro-Grid Research/Generator output data/Port a Piment/Full year time series/papim_power_tot.csv')
ldc = power_to_ldc(power)
plots(pap, ldc, 'pap')
```
