# Conductivity and Insulation Volume

This notebook calculates the thermal conductance for a spherical shell of insulation.  I'm doing this to explore the economic and technical limits of increasing the amount of insulation for refrigerators.  

```python
%%capture
%pylab inline
```

```python
# inputs

# value for polyurethane foam
k = 0.02
# cost for polyurethane spray foam 
insulation_USDpm3 = 1. / 144. / 2.54**3 * 10.**6 
insulation_USD_per_m3 = 400.

print('insulation cost (USD per cubic meter) =', insulation_USDpm3)

# we set the radii of the shell and report the interior volume
r1 = 0.288
r2 = linspace(1.05 * r1, 3 * r1, 100)
interior_volume = 4 / 3 * pi * r1**3

print('interior volume (cubic meter) =', interior_volume)

# computes the thermal resistance and conductance
thermal_resistance = (1/r1 - 1/r2) / 4 / pi / k
thermal_conductance = 1 / thermal_resistance

thermal_conductance_limit = 4 * pi * k * r1

print('limit of thermal conductance for infinite insulation (W/K) =', thermal_conductance_limit)
# insulation cost
volume_insulation = (r2**3 - r1**3) * 4 * pi / 3
cost_insulation = volume_insulation * insulation_USDpm3 * 0.33
```

```python
print('interior volume = ', interior_volume)
plot(r2-r1, thermal_conductance)
xlabel('insulation thickness (m)')
ylabel('thermal conductivity (W/K)')
grid(); show()
```

```python
print('interior volume = ', interior_volume)
plot(volume_insulation, thermal_conductance)
xlabel('volume of insulation (m$^3$)'); ylim((0, 2.0))
ylabel('Thermal Conductance (W/K)'); xlim((0, 1.0))
grid(); show()
```

```python
print('interior volume = ', interior_volume)
plot(volume_insulation, thermal_conductance / thermal_conductance_limit)
xlabel('volume of insulation (m$^3$)')
ylabel('factor above thick insulation conductivity limit'); ylim((0, 10))
grid(); show()
```

There are two practical limits to the amount of insulation.  The first is the cost of the insulation.  At a few hundred, USD per cubic meter, this cost can be significant.  The second practical limit is the thickness.  Customers may not accept a large refrigerator with a small storage space.  There may also be practical aspects of creating a door and a seal with such a thick insulation.  It is plausible also that leakage paths or other practical issues could increase the thermal conductance rendering the extra insulation useless.

## Comparison to currently available units

I estimate that inexpensive 100 liter refrigerators have a conductance of 1.1 W/K and highly efficient models have 0.5 W/K.  Technically, it seems feasible to have a better insulated thermal envelope, considering only the insulation thickness.
