# How does lifetime refrigeration cost vary with increasing energy cost?

A fairly simple calculation I want to carry out is the optimum lifecycle cost of refrigeration for a given volume and a given cost of energy.  As energy cost rises, it should be economical to raise the amount of insulation.  The amount of insulation will lead to a minimum life-cycle cost.  I'm worried that this should be obviously linear but I can't see why yet.

For a given climate, interior temperature, and envelope conductance the heat to be removed is the same, so the daily energy cost will increase linearly with the per kWh cost of energy.  As the energy cost increases, it becomes economical to add insulation.

I can adapt the calculations from 18 Oct 2013 to draw out this plot.



```python
%%capture
%pylab inline

#assumptions
# polyurethane foam
k = 0.02
# cost for polyurethane spray foam 
insulation_USD_per_m3 = 400
# cost of capital
CRF = 0.3
# refrigeration coefficient of performance
COP = 2.0
joules_per_kWh = 3.6e6 
seconds_per_year = 3.15e7
energy_cost = 0.15
temperature_difference_K = 20

# 100 liter volume refrigerator
r1 = 0.288
r2 = linspace(1.05 * r1, 10.0 * r1, 1000)
interior_volume = 4 / 3 * pi * r1**3
```

```python
def get_optimal_life_cycle_cost(energy_cost=energy_cost):
    
    # computes the thermal resistance and conductance
    thermal_resistance = (1/r1 - 1/r2) / 4 / pi / k
    thermal_conductance = 1 / thermal_resistance
    power = thermal_conductance * temperature_difference_K
    
    # insulation cost
    volume_insulation = (r2**3 - r1**3) * 4 * pi / 3
    cost_insulation = volume_insulation * insulation_USD_per_m3
    
    # yearly costs
    capital_cost = cost_insulation * CRF
    operational_cost = power * seconds_per_year / joules_per_kWh / COP * energy_cost 
    life_cycle_cost = capital_cost + operational_cost
    
    # quick and dirty optimization
    # if grid is coarse, optimum will be missed
    # also, if r2 doesn't encompass whole range, bad result
    return min(life_cycle_cost), (r2-r1)[argmin(life_cycle_cost)]
```

```python
# this is klunky, all parameters should be in dict?
energy_cost = linspace(0.01, 5.00, 50)
insulation_USD_per_m3 = 400.
lcc1 = [get_optimal_life_cycle_cost(ec)[0] for ec in energy_cost]
insulation_USD_per_m3 = 200.
lcc2 = [get_optimal_life_cycle_cost(ec)[0] for ec in energy_cost]

plot(energy_cost, lcc1, label='expensive insulation')
plot(energy_cost, lcc2, label='cheap insulation')
ylabel('Yearly cost of refrigeration (USD/year)')
xlabel('Cost per kWh of electricity (USD/kWh)')
grid()
legend()
show()
```

There appears to be an initial regime with sub-linear behavior and then it approaches linear behavior.  This calculation does not consider possible savings from a smaller, cheaper compressor.

```python
# this is klunky, all parameters should be in dict?
energy_cost = linspace(0.01, 5.00, 50)
insulation_USD_per_m3 = 400.
size1 = [get_optimal_life_cycle_cost(ec)[1] for ec in energy_cost]
insulation_USD_per_m3 = 200.
size2 = [get_optimal_life_cycle_cost(ec)[1] for ec in energy_cost]

plot(energy_cost, size1, label='expensive insulation')
plot(energy_cost, size2, label='cheap insulation')
ylabel('Thickness of walls (meters)')
xlabel('Cost per kWh of electricity (USD/kWh)')
grid()
legend()
show()
```

Cheap insulation leads to lower yearly costs but can increase the wall thickness of a refrigerator.
