# Price Duration Curve

This notebook generates graphs of a price duration curve corresponding to demand data from microgrids.  From the generator specification of fuel use rate in liters per hour and percent of rated load output power, we create a linear model of fuel use.  Since much of the observed demand is below the lowest point on the specification, we use a linear extrapolation.  From this we calculate the liters per kWh and calculate a price duration curve which shows the per kWh price for power against the duration. 

```python
%pylab inline

class Generator:
    '''
    simulates a generator with utilities to input fuel use given an output load
    
    efficiency curve is a dictionary with the fuel use per hour in liters and the kW load as two lists or arrays
    
    '''
    def __init__(self, efficiency_curve):
        # store efficiency curve as variable
        self.efficiency_curve = efficiency_curve
        # perform linear fit and store polynomial coefficients for linear fit
        self.fuel_use_coeff = self.fuel_rate_fit()
    
    def fuel_rate_fit(self):
        # perform a linear fit for the fuel use rate specification and return coefficients
        p = polyfit(self.efficiency_curve['output_power'],
                    self.efficiency_curve['fuel_use'],
                    1)
        return p

    def get_fuel_use(self):
        # convenience method to get data sheet fuel use for plotting
        return self.efficiency_curve['fuel_use']
    
    def get_output_power(self):
        # convenience method to get data sheet output power for plotting
        return self.efficiency_curve['output_power']
    
    def fuel_at_load(self, load):
        # use the linear model coefficients from fuel_rate_fit() to estimate fuel use for load
        fuel = polyval(self.fuel_use_coeff, load)
        return fuel
    
    def efficacy_at_load(self, load):
        # returns the fuel use in liters per kWh of electricity delivered at a given load
        return self.fuel_at_load(load) / load

def plots(pap, ldc):
    
    # get the maximum and minimum power observed in our demand data
    max_power = nanmax(ldc)
    min_power = nanmin(ldc)

    # create a linearly spaced range for plotting based on this range
    power_range = linspace(min_power, max_power, 10)
    
    # plot the fuel consumption per hour of the generator according to the specification
    # also plot the inferred fuel consumption over the observed range of use
    plot(pap.get_output_power(), pap.get_fuel_use(), 'bx', label='Specification')
    plot(power_range, pap.fuel_at_load(power_range), 'k', label='Range of Use')
    title('Hourly Fuel Use')
    xlabel('Load (kW)')
    ylabel('Fuel Consumption (liters/hour)')
    legend(loc='best')
    show()
    
    # plot the fuel consumption per kWh
    plot(pap.get_output_power(), pap.efficacy_at_load(pap.get_output_power()), 'bx', label='Specification')
    plot(power_range, pap.efficacy_at_load(power_range), 'k', label='Range of Use')
    title('Fuel Use per Unit Electrical Energy Delivered')
    xlabel('Load (kW)')
    ylabel('Fuel Consumption (liters/kWh)')
    ylim((0.2, 0.5))
    legend(loc='best')
    show()
    
    # plot the load duration curve
    plot(ldc)
    title('Load Duration Curve')
    xlabel('counts')
    ylabel('Power kW')
    show()
    
    # plot the price per kWh duration curve
    plot(pap.efficacy_at_load(ldc))
    title('Price Duration Curve')
    xlabel('Counts')
    ylabel('Electricity Cost (USD/kWh)')
    show()
```

# Coteaux Kohler Generator

- from data sheet in haiti folder
- coteaux kohler
- percent load, standby lph, prime lph
- 100%, 44.3, 40.6
- 75%, 35.1, 32.3
- 50%, 26.3, 24.0
- 25%, 16.2, 14.4
- no load consumption not given
- from report: rated active power 125kW
- Manufactured 1994. 


```python
# create a generator object using the data from the Coteaux spec sheet
power_kW = 125.
efficiency_curve = {'output_power': array([0.25, 0.5, 0.75, 1.0]) * power_kW,
                    'fuel_use' : array([14.4, 24.0, 32.3, 40.6])}
coteaux = Generator(efficiency_curve)


power = loadtxt('/Users/dsoto/Dropbox/Haiti Micro-Grid Research/Generator output data/Coteaux/Full year time series/coteaux_power_tot.csv')
# convert raw time series to load duration curve by sorting
ldc = numpy.sort(power)
# truncate empty values (done manually)
ldc = ldc[339000:]

# generate plots
plots(coteaux, ldc)
```
