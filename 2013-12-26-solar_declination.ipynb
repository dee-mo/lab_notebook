# solar equations

i'm going to plot the solar declination but with dates for the x axis.

```python
%pylab inline
max_declination = 23.45
days_per_year = 365
equinox_offset = 81
def declination(n):
    return max_declination * sin(2 * pi * (n - 81) / days_per_year)
```

```python
days = range(1, 366)
dec = [declination(n) for n in days]
plot(days,dec)
```

```python
# construct datetimes from range of days
# use these datetimes as x axis labels
import datetime as dt
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
```

```python
dates = [dt.datetime.strptime(str(n), '%j') for n in days]
```

```python
fig, ax = plt.subplots()
# rock the formatters
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b'))
ax.plot(dates, dec)
#fig.autofmt_xdate()
ax.set_title('Solar Declination')
ax.set_ylabel('Degrees North of equator')
ax.grid()
plt.show()
```
