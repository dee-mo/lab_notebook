# Calculation of Net Present Value

Using the python numerical libraries.

```python
%pylab inline
```

Create a series of payments and plot on graph.

```python
years = array([1, 2, 3])
payments = array([100., 200., 300.])
plot(years, payments, 'ko')
ylim((0, 400)); ylabel('Future Payment (USD)')
xlim((0, 4)); xlabel('Time (years)')
title('Payment Stream')
grid()
```

Calculate the net present value using a library function.

```python
discount_rate = 0.1
npv(discount_rate, payments)
```
