## IRR of a solar panel

As the simplest exercise, let us make the assumption that we can purchase a solar panel and then sell all the electricity it produces in a year at a given rate.  From this, we can calculate the IRR of the panel as an investment.

```python
def default_inputs():
    # create default values for solar investment
    d = {}
    d['solar_panel_specific_cost_USDperwatt'] = 1.0
    d['site_peak_sun_hours'] = 5.0
    d['value_of_energy_USDperkWh'] = 0.10
    d['life_of_panel'] = 20
    d['panel_size_watts'] = 1.
    return d

def set_data(md):
    # allow for customization of parameters
    d = default_inputs()
    for m in md:
        d[m] = md[m]
    return d

def calc_derived(d):
    # calculate derived parameters and add to dictionary
    d['yearly_energy'] = 365 * d['site_peak_sun_hours'] * d['panel_size_watts'] / 1000.
    d['panel_cost'] = d['panel_size_watts'] * d['solar_panel_specific_cost_USDperwatt']
    d['yearly_income'] = d['yearly_energy'] * d['value_of_energy_USDperkWh']
    return d

def cash_flow(d):
    return np.hstack(([-d['panel_cost']], [d['yearly_income']] * d['life_of_panel']))

def solar_irr(md):
    d = set_data(md)
    d = calc_derived(d)
    d['cash_flow'] = cash_flow(d)
    d['irr'] = np.irr(d['cash_flow'])
    return d

def grid_irr(panel_cost, electricity_price):
    # convenience function to be vectorized to allow for plotting
    md = {'solar_panel_specific_cost_USDperwatt':panel_cost,
          'value_of_energy_USDperkWh':electricity_price}  
    return solar_irr(md)['irr']

vgrid = np.vectorize(grid_irr)
```

```python
md = {'solar_panel_specific_cost_USDperwatt':2.1,
      'value_of_energy_USDperkWh':0.10}
results = solar_irr(md)
print('irr = ', results['irr'])
plot(results['cash_flow'], 'bd')
grid()
```
