# Energy Balance Model

My goal is to create a simple, extensible model for energy balance analysis and optimization.  The idea is to have components or objects that conform to a simple protocol having energy input, energy output, and parameters that affect the performance of the object.  These objects would be batteries, solar panels, insulation, or others.  These objects would then be placed in a directed graph representing the energy flow.  This model could then be iterated by timestep to create a timeseries.  

The model would be used for techno-economic optimizations that don't require precise engineering detail to obtain reasonable estimates.  

I'd like to build this for the case of a PV powered refrigerator.  The simulation would determine the temperature time series and cost required to achieve it.

## Energy conversion devices

### Solar panel
- input - incident radiation
- output - electrical energy
- parameters - efficiency, temperature performance, size, cost, etc

### Insulation
- input - heat from outside
- output - heat to inside
- parameters - thickness, conductivity, cost, amount

### Refrigeration system
- input - electrical energy input
- output - thermal energy removed
- parameters - coefficient of performance, cost, size

### Interior volume
- change in mass, temperature of incoming mass
- total interior mass and heat capacity
- ambient heat entering
- heat pump heat leaving
- overall temperature

Each of these pieces will be able to take the energy output of another as the energy input.

## Energy sources

### Grid electricity
- output - electricity delivered
- parameters - maximum output, cost per unit energy

### Sunlight
- output - radiation delivered
- parameters - maximum output, cost per unit energy

### Thermal reservoir 
- output - heat
- parameters - maximum heat flux, cost per unit energy



## Economic optimization

The purpose of this is to find economically optimal systems.  The objective function could be a minimum acceptable level of service defined in terms of light delivered, hours of availabililty, or temperature range.  The sizes of the components could then be converted to levelized costs and reported.  

The brute force way to do this is to calculate a time series, extract the objective function from the time series, and then use that in an optimization function.  This will likely have performance issues but could be more approachable by a wide audience than casting it as a linear programming problem.  Linearization would avoid the inevitable convergence problems with a brute force approach. 


## Time Steps

This should be a basic forward time step.  The parameters will be determined for $t_i$ and then the energy flows will be calculated for $t_{i+1}$.  Based on these energy flows, the parameters can be determined for $t_{i+1}$.  

## Control system

The compressor could either run on a control, or simply run at a rate based on the incident solar radiation.  This runs the risk of over-cooling the interior on sunny and cool days.  Another option is to install a controller to turn on and off the compressor.  The problem here is that valuable solar electricity will be unused.
