```python
import pandas as pd
%pylab inline
import scipy
from scipy.constants.constants import C2F
```

```python
!head '/Users/dsoto/Dropbox/RCH/RCH12-hobo.csv'
```

```python
temp_data = pd.read_csv('/Users/dsoto/Dropbox/RCH/RCH12-hobo.csv', 
                        skiprows=1, 
                        usecols=[1,2], 
                        parse_dates=True,
                        index_col=0)
```

```python
temp_data['fahr'] = C2F(temp_data['Temp, °C (LGR S/N: 1257433, SEN S/N: 1257433)'])
```

```python
temp_data['fahr'].plot()
ylabel('Air Supply Temperature (F)')
title('Air Supply Rachel Carson Hall 12 (HOBO)')
show()
```
