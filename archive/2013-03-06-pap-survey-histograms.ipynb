```python
%pylab inline
import pandas as pd
xf = pd.ExcelFile('Load survey results 20130214.xlsx')
survey = xf.parse('Microgrid_survey_20120903', index_col=None)
```

```python
# subselect Port a Piment respondents and place in data frame
pap = survey[survey['city']=='P a piment']
```

```python
import matplotlib.pyplot as plt
for col in pap.columns:
    if col and '_num' in col:
        pap[col].hist(normed=True, bins=range(10))
        plt.title(col)
        plt.show()
```
