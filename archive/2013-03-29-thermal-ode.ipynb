# using ODE to solve the temperature of a heated body

This is a warm-up exercise for using the ODE solvers in scipy to get temperature profiles for lumped mass systems like a refrigerator or a crockpot.

For a simple system, we can write an energy balance equation and use it to determine the derivative of the temperature with respect to time for the system.  We can then numerically integrate the system to find the temperature behavior.

We start with a simple system with a known heat capacity and thermal envelope cooling to the ambient temperature.  The change in stored energy is equal to the energy lost through conduction.

$$ \dot{E_s} = -k (T - T_{amb} ) = 1/c dT/dt $$ 

$$ \dot{E_s} = P -k (T - T_{amb} ) = 1/c dT/dt $$ 

# Data and simulation

I used the small crockpot filled with 500 ml of water on low heat and am recording the temperature of the water at approximately 10 minute intervals.  At 4.8 J/g/K, the water should have a heat capacity of about 2400 J/K.  I'm assuming about the same for the ceramic.  The electrical power demanded by the pot is 72 watts which I'm assuming is completely delivered to the pot.

```python
%pylab inline
import scipy.integrate as spi
```

# Cooling Run

From the cooling run, we can fit the ratio of the thermal capacity to the thermal conductivity.  

```python
# raw data from cooling run
data = array([[0, 95.5], [6, 91.8], [15, 84.5], [28, 75.6], [45, 67.4], [67, 58.3], [93, 50.8], [116, 45.7]
             ])
data[:,0] *= 60

def dTdt(T,t):
    power = 0.0
    Tamb = 20
    K = 166 * 1e-6
    C = 1
    out = (power - K * (T - Tamb)) / C
    return out

# use data samples for time base
t = data[:, 0]
yobs = data[:, 1]

# use ODE solver to 
Tinit = 95.5
ymod = spi.odeint(dTdt, Tinit, t)

y = spi.odeint(dTdt, 95.5, t)

# plot simulation
plot(t, ymod, 'bx', label='sim')

# plot measured data
plot(t, yobs, 'kx', label='data')
title('free cooling')

legend()
print('squared error =', sum((ymod.T - yobs)**2))

show()

plot(t, (y.T - data[:, 1])[0], 'x')
title('residuals')
show()
```

# Heating run

For the heating run, I believe we can separate the conductivity and the heat capacity since the power is non-zero.

```python
# raw data from heating run
data = array([[0, 20.5], [10, 25.5], [23, 36.0], [30, 41.5], [42, 50], [50, 55.8], 
              [72,68.7], [110, 83.7], [145, 90.8], [163, 92.4], [196, 95.5]  
             ]) 
data[:,0] *= 60

# derivative of temperature function
def dTdt(T,t):
    power = 75
    Tamb = 17
    C = 4800
    K = 168 * 1e-6 * C
    out = (power - K * (T - Tamb)) / C
    return out

# use data samples for time base
t = data[:, 0]
yobs = data[:, 1]

# use ODE solver to 
Tinit = 20
ymod = spi.odeint(dTdt, Tinit, t)

# plot simulation
plot(t, ymod, label='sim')

# plot measured data
plot(t, yobs, label='data')

legend()
print('squared error =', sum((ymod.T - yobs)**2))
print('thermal conductance =', 4800 * 168 * 1e-6)
```

From this I estimate that the heat capacity of the water plus pot is about twice what I estimate for the water and that the thermal conductivity is about 0.8 J/K.  I'll have to read up on convective heat loss and see if this makes sense.
