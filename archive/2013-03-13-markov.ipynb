## Markov chain simulations

Here I make some preliminary attempts at simulating a markov chain.  There may be an existing library for this but the exercise will be worth it, I'm sure.  This prototype is a simple two-state markov chain that could represent a device like a light bulb, or refrigerator compressor that will only have on and off states.  At each time step, the probability of the next state is calculated from the transition matrix and the previous state.

The transition probabilities in this two state matrix are from on to on, from on to off, from off to on, and from off to off.  

$
\textrm{transition matrix} =
\begin{pmatrix}
p(off|off) & p(off|on) \\\
p(on|off) & p(on|on)
\end{pmatrix}$

The state matrices are column probability vectors that have the form

$\textrm{state} = 
\begin{pmatrix}
p(off)\\\
p(on)
\end{pmatrix}$

The multiplication yields the next state

$
\textrm{next state} =
\begin{pmatrix}
p(off|off)p(off) + p(off|on)p(on) \\\
p(on|off)p(off) +  p(on|on)p(on)
\end{pmatrix}$

In this simulation, we will use this probability distribution to choose the next state as either (0,1) or (1,0).

From this we can form a sequence of probability column vectors in a 2xN matrix.  We can also represent the power in each state as a 1x2 vector for a two-state on-off system.  Multiplying these matrices will yield a row vector time series of the power consumed by the device.

The power matrix is
$P = \begin{pmatrix}
power(off) & power(on)
\end{pmatrix}
$


```python
%pylab inline

'''
this function runs a markov chain simulation for a two state matrix

inputs:
transition_matrix: 2x2 matrix for a two-state markov chain
initial_state: defines whether appliance is initially on or off
num_steps: number of time steps in simulation
plot_result: boolean controlling output plot

outputs:
row vector of observed power consumption by appliance over time steps
'''
def run_sim(transition_matrix, initial_state=array([0, 1]), num_steps=100, plot_result=False):
    # create state sequence matrix
    state_sequence = zeros((2, num_steps))
    # set initial state (default is on)
    state_sequence[:, 0] = initial_state
    # loop over array and calculate probabilities and states
    for i in range(1, num_steps):
        # calculate probability vector for next state
        next_state_probability = transition_matrix.dot(state_sequence[:, i-1])
        # choose next state based on probability vector
        on = random.choice([0, 1], p=next_state_probability)
        # insert state into sequence based on on/off value chosen by random
        if on == 1:
            state_sequence[:, i] = [0, 1]
        elif on == 0:
            state_sequence[:, i] = [1, 0]
    
    # define power for each state
    power_off = 0.
    power_on = 10.
    powers = array([[power_off, power_on]])
    # calculate consumed power by multiplying power and state sequence vectors
    observed_load = powers.dot(state_sequence)
    # visualize
    if plot_result:
        plot(observed_load[0], 'kx-')
        title('Single Appliance Power Demand')
        xlabel('Time Index')
        ylabel('Watts')
        show()
    return observed_load[0]
```

## Run an instance of the simulation for several identical appliances

```python
# define transition probability matrix
on_to_on = 0.7
on_to_off = 1.0 - on_to_on
off_to_off = 0.7
off_to_on = 1.0 - off_to_off
transition_matrix = array([[off_to_off, on_to_off], [off_to_on, on_to_on]])

# define number of time steps and appliances
num_steps = 500
num_appliances = 200

# create accumulator to store power for all appliances
total_power = zeros(num_steps)

# manually seed random number generator 
numpy.random.seed(0)

# loop over appliances creating random markov chain sample and accumulate
# power in a row vector, total
for i in range(num_appliances):
    total_power += run_sim(transition_matrix, num_steps=num_steps, plot_result=False)
```

## Single time series of appliance load

```python
run_sim(transition_matrix, plot_result=True)
show()
```

## Aggregated demand over all appliances

```python
plot(total_power)
title('Aggregated Power Consumed')
xlabel('Time Index')
ylabel('Average Power per Appliance')
ylim(ymin=0)
show()
```

## Histogram of observed power levels

```python
hist(total_power, bins=50)
title('Distribution of Power')
xlim(xmin=0)
xlabel('Average Power per Appliance')
ylabel('Frequency')
show()
```

## Frequency content of overall load

```python
semilogy(fft.ifftshift(abs(fft.fft(total_power))))
xlabel('Wavenumber')
title('Spectrum of Frequency Content')
show()
```
