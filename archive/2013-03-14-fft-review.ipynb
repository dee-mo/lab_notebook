# A refresher on FFTs

$X_k = \sum_{n=0}^{N-1} x_n \exp(-i 2 \pi k n / N) $

$k$ is the frequency component index ranging from $k = 0$ to $N-1$

$N$ is the number of samples in the window.

$n$ indexes the samples

For $k=0$, this is the DC component.  

For $k=1$, is one cycle over the entire sample window.  So the angular frequency is 1/N 1/min

For $k=2$, there are two cycles over the sample window.  The frequency is 2/N 1/min$

For $k=N/2$ the frequency is 1/2 1/min.  Anything above this is above nyquist and will be aliased frequencies.

The higher $k$ is the higher frequency.

$cos(2 \pi k t)$

For purely real data, $X_{N-k} = X^{*}_{k}$.

```python
%pylab inline
```

Create timebase and then generate signal with multiple frequency components and plot.

```python
N = 200
t_max_minutes = 10
t = linspace(0, t_max_minutes, N)
dt = t_max_minutes / N

f1 = 1.0
a1 = 1.0
f2 = 5.0
a2 = 0.5
y = a1*sin(f1 * 2 * pi * t) + a2*cos(f2 * 2 * pi * t) + 0.3 * rand(N)

plot(t, y)
xlabel('Time (min)')
```

## Unshifted raw FFT

```python
ft = abs(fft.fft(y))
plot(ft)
```

## Shifted raw FFT

Zero frequency is in the middle of the x-axis.

```python
plot(fft.fftshift(ft))
```

## Half frequency space with proper scale frequency axis

```python
f = arange(N) / N / dt
plot(f[0:N/2],ft[0:N/2])
xlabel('Frequency (1/min)')
```
