```python
%%octave

num_users = 100;
time_steps = 100;

appliance_ownership = simulate_appliance_ownership(num_users);

load_data = generate_time_series(appliance_ownership, time_steps);

% collapse array to time dimension and plot
figure
plot(sum(sum(load_data))(:))
title('overall load')

% plot single customer
customer_index = 10;
single_customer_data = sum(load_data, 2)(customer_index, :, :)(:);
figure
plot(single_customer_data)
title('single customer')

% plot single appliance type
appliance_index = 2;
appliance_type_data = sum(load_data, 1)(:, appliance_index, :)(:);
figure
plot(appliance_type_data)
title('single appliance type')
```
