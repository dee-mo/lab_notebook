```python
%pylab inline
import matplotlib as mpl
```

```python
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['font.size'] = 14
mpl.rcParams['grid.linestyle'] = '-'
mpl.rcParams['grid.alpha'] = 0.2
mpl.rcParams['legend.fontsize'] = 12
```

We set up three waveforms with an amplitude of 120 and the 60 Hz frequency.

```python
A = 1
omega = 377

t = linspace(0, 1/60, 100)

Va = A * cos(omega * t)
Vb = A * cos(omega * t + 2 * pi / 3)
Vc = A * cos(omega * t + 4 * pi / 3)
```

Here we plot these phases.  Notice the symmetry in the waveforms.

```python
plot(t, Va)
plot(t, Vb)
plot(t, Vc)
xticks(linspace(0, 1/60., 5))
grid()
show()
```

If we take the difference between any of these phases, we see that we get waveforms at a higher voltage.

```python
plot(t, Va - Vb)
plot(t, Vb - Vc)
plot(t, Vc - Va)
xticks(linspace(0, 1/60., 5))
grid()
show()
```

```python
fig, ax = subplots(1,1)
ax.plot(t, Va, label='Va phase')
ax.plot(t, Vc, label='Vc phase')
ax.plot(t, Va - Vc, label='Vac line')
grid()
legend(loc='best')
xlabel('Time (sec)')
ylabel('Voltage')
ax.set_ylim((-2.5, 2.5))
ax.set_yticks((-1.73, -1, 0, 1, 1.73))
ax.set_yticklabels((r'$-\sqrt{3}=-1.73$', r'$-1.0$', r'$0$', r'$1.0$', r'$\sqrt{3}=1.73$'))
ax.set_xticks((0.0, 1/120, 1/60))
ax.set_xticklabels((r'$0$', r'$\frac{1}{120}$', r'$\frac{1}{60}$')) 
title('Line and Phase Voltages')
savefig('line-and-phase-voltages.pdf')
savefig('line-and-phase-voltages.png')
show()
```
