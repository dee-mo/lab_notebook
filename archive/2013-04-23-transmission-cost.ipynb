```python
%pylab inline

wire_length = linspace(0, 500, 100)
kWh_per_year = linspace(10, 200, 100)

wl, kpy = meshgrid(wire_length, kWh_per_year)

cost_per_meter = 1.0
capital_recovery_factor = 0.15

lifetime_cost_kWh = wl * cost_per_meter * capital_recovery_factor / kpy

cp = contour(wl, kpy, lifetime_cost_kWh, colors='#aaaaaa', 
                                       levels=[1, 0.5, 0.25, 0.1], linewidths=0.5)
clabel(cp, inline=1)
xlabel('length of wire')
ylabel('kWh per year')
title('lifetime transmission cost per kWh\ncost per meter (USD)=%1.1f\ncapital recovery factor=%0.2f'
 % (cost_per_meter, capital_recovery_factor))
#show()


text(100, 100, 'queen 003 customers?')
```
