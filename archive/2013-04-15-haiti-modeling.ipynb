# Haiti consumer modeling approaches

On Friday, Dan Schnitzer and I met and discussed the Markov modeling I've been doing.  We are both concerned about the distribution of on and off times that results from the Markov chain since the most frequent on or off time is 1 period, which does not accurately capture the behavior of many appliances, especially lights or televisions.

We are considering an approach where the start time and either the end time or the duration are drawn from probability distributions.  If we take the case of evening lighting, we may assume that the lights will turn on near sunset and that the lights will be extinguished around the usual bedtime of the villagers.  

Here is the basic idea, in python.

```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt

day_length = 24 * 60

start = 500
start_deviation = 15

start_sample = np.random.normal(loc=start, scale=start_deviation, size=1000)
plt.hist(start_sample, bins=50)
plt.title('start time distribution')
plt.xlim((0, day_length))
plt.show()

end = 1000
end_deviation = 60
end_sample = np.random.normal(loc=end, scale=end_deviation, size=1000)

plt.hist(end_sample, bins=50)
plt.title('end time distribution')
plt.xlim((0, day_length))
plt.show()
```

From these distributions, we can construct a time series.  Start by assuming a certain total length (minutes or hours in a day, perhaps).  Fill in zeros before start time, fill in ones from start time to end time, and fill in zeros afterward.


```python


accum = np.zeros(day_length)

for i in range(1000):
    ts = np.zeros(day_length)
    
    start_time = np.random.normal(loc=start, scale=start_deviation, size=1)
    end_time = np.random.normal(loc=end, scale=end_deviation, size=1)
    
    start_time = int(start_time)
    end_time = int(end_time)
    ts[start_time:end_time] = 1.0
    accum += ts

plot(accum)
```

Since we have chosen a fairly sharp start time distribution and a fairly broad off time distribution, we see a quick rise in the load around the start time and then a slower decline around the end time.

This very simple technique is well suited for loads that will likely be on for one continuous session each evening.  You could extend this model to something that sequences on and off, and use on and off duration distributions.  For most consumer devices, I think this will only be accurate over certain times of day.
