# Display of acquired data

This notebook allows the display of data stored in CSV by a LabJack device.

```python
%pylab inline
```

```python
# perform smoothing of data
# show energy delivered
```

```python
data = loadtxt('labjack.txt', delimiter=',')
fig, ax = subplots(2,1)
current_calibration = 1.0
ax[0].plot(data[:,0], data[:,1])
ax[0].set_ylabel('Battery Voltage (V)')
ax[1].plot(data[:,0], data[:,2]*current_calibration)
ax[1].set_ylabel('Shunt Voltage (V)')
ax[1].set_xlabel('Time (sec)')
show()
```

```python
data[:,2].mean()
```

```python
# gets quantization levels by taking unique differences
unique(diff(data[:,2]))
```
