```python
datafile = 'battery-charge-2013-11-11.txt'
data = loadtxt(datafile, delimiter=',')
fig, ax = subplots(2,1)
current_calibration = 0.2 / 0.034
ax[0].plot(data[:,0], data[:,1])
ax[0].set_ylabel('Voltage (V)')
ax[1].plot(data[:,0], data[:,2]*current_calibration)
ax[1].set_ylabel('Current (A)')
ax[1].set_xlabel('Time (sec)')
show()
```

```python
plot(data[:,1] * data[:,2])
title('Energy Delivered to Battery')
```
