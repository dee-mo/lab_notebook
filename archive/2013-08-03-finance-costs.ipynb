# Contour plots of cost of capital

I don't find the formulas for a montly payment intuitive.  To try to gain some intuition, I plot the payment for a given principal given different annual percentage rates and loan lengths.  

```python
%pylab inline
import numpy as np
```

## Monthly Payments

The monthly payment plot is useful for thinking about vehicle loans.  The APR is adjusted by dividing by 12 to compound over a monthly period.

```python
principal = 4200

# create variable grids
periods = linspace(24, 72, 5)
apr = linspace(0.001, 0.10, 10)
PER, APR = np.meshgrid(periods, apr)

# compute payment
payment = -np.pmt(APR/12, PER, principal)

# plot and annotate
#CRF_plot = contour(PER, APR, payment, colors='#aaaaaa', levels=[20, 25, 30, 40])
CRF_plot = contour(PER, APR, payment, colors='#aaaaaa', levels=None)
clabel(CRF_plot, inline=1)
xlabel('Number of months')
ylabel('APR')
title('Monthly Payment on principal of {}'.format(principal))
grid()
```

## Yearly payments

The yearly payments are useful for thinking about a solar panel or house loan or other large asset.

```python
periods = linspace(1, 20, 10)
apr = linspace(0.001, 0.10, 10)
P, A = np.meshgrid(periods, apr)
principal = 1
payment = -np.pmt(A, P, principal)
CRF_plot = contour(P, A, payment, colors='#aaaaaa', levels=[0.1, 0.15, 0.2, 0.3, 0.5])
clabel(CRF_plot, inline=1)
xlabel('Number of years')
ylabel('APR')
title('Yearly Payment (CRF)')
grid()
```
