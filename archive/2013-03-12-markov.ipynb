## Markov chain simulations
Here I make some preliminary attempts at simulating a markov chain.

Here I prototype a simple two state markov chain that could represent a device like a light bulb, or refrigerator compressor that will only have on and off states.  At each time step, the probability of the next state is calculated from the transition matrix and the previous state.

$a_{ik} = b_{ij} c_{jk}$
The transition matrix is $b$.  The initial state is $c$ and the next state is $a$.

The transition probabilities in this two state matrix are from on to on, from on to off, from off to on, and from off to off.  

$b_{11} = p(on|on)$

$b_{12} = p(on|off)$

$b_{21} = p(off|on)$

$b_{22} = p(off|off)$

The state matrices have the form

$c_{11} = p(off)$

$c_{21} = p(on)$

All these columns should add to one since they are probability vectors.

From this we can form a sequence of probability vectors in a 2xN matrix.  We can also represent the power in each state as a 1x2 vector for a two-state on-off system.  Multiplying these matrices will yield a row vector time series of the power consumed by the device.



```python
%pylab inline
def run_sim(transition_matrix, initial_state=array([1, 0]), N=100):
    # create state sequence matrix
    state_sequence = zeros((2, N))
    # set initial state (default is off)
    state_sequence[:, 0] = initial_state
    # loop over array and calculate probabilities and states
    for i in range(1, N):
        # calculate probability vector for next state
        next_state_probability = transition_matrix.dot(state_sequence[:,i-1])
        # choose next state based on probability vector
        on = random.choice([0,1], p=next_state_probability)
        # insert state into sequence based on on/off value chosen by random
        if on == 1:
            state_sequence[:, i] = [0, 1]
        elif on == 0:
            state_sequence[:, i] = [1, 0]
    
    # define power for each state
    powers = array([[10, 0]])
    # calculate consumed power by multiplying power and state sequence vectors
    emission = powers.dot(state_sequence)
    # visualize
    plot(emission[0], 'kx-')
    show()
    return emission[0]
```

## Mostly on
If I assign a high probability to staying on and a smaller probability to staying off, I get a power time series of mostly on.

```python
on_to_on = 0.9
on_to_off = 1.0 - on_to_on
off_to_off = 0.1
off_to_on = 1.0 - off_to_off
transition_matrix = array([[on_to_on, off_to_on], [on_to_off, off_to_off]])
for i in range(4):
    run_sim(transition_matrix)
```

## Bistable
If you assign high probabilities to both staying on and staying off, you get longer dwell times in both states.

```python
on_to_on = 0.9
on_to_off = 1.0 - on_to_on
off_to_off = 0.9
off_to_on = 1.0 - off_to_off
transition_matrix = array([[on_to_on, off_to_on], [on_to_off, off_to_off]])
for i in range(4):
    run_sim(transition_matrix)
```
