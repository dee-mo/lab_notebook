# Recent CAISO Solar Thermal Production

Scrapes the daily renewables watch page for recent data.  I poked around for a better API but couldn't find one.

```python
%pylab inline

def get_df_for_date(date):
    url = 'http://content.caiso.com/green/renewrpt/' + date + '_DailyRenewablesWatch.txt'
    df = pd.read_table(url, sep='\t+', skiprows=1, nrows=24)
    return df

def plot_solar_thermal(date):
    df = get_df_for_date(date)
    tsum = df['SOLAR THERMAL'].sum()
    plot_title = '{}\nCAISO Solar Thermal Output (MW)\nTotal for day = {} MWh'.format(date, tsum)
    df['SOLAR THERMAL'].plot(title=plot_title)
    show()
```

```python
for date in ['20140208','20140209','20140210','20140211','20140212', '20140213']:
    plot_solar_thermal(date)
```
